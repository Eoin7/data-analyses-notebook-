```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

background = 236
file_name = ["0.txt", "0.5.txt", "1.txt", "2.txt", "3.txt", "5.txt"]

# x values - concentration
x = np.array([0, 0.5, 1, 2, 3, 5])

fig = plt.figure(figsize=(12, 10))
global_min = float('inf')
global_max = float('-inf')

# Load data and find global min/max
for fname in file_name:
    pixel_data = np.loadtxt(fname)
    pixel_array = pixel_data - background
    global_min = min(global_min, pixel_array.min())
    global_max = max(global_max, pixel_array.max())

y = []

# Load data, adjust and display images
for i in range(6):
    pixel_data = np.loadtxt(file_name[i])
    pixel_array = pixel_data - background
    ax = fig.add_subplot(2, 3, i+1)
    im = ax.imshow(pixel_array, cmap='inferno', aspect='equal', vmin=global_min, vmax=global_max)
    ax.set_title(f"{x[i]} mm pixel intensity {int(pixel_array[269, 271])}")  # Set the title to the name of the file

    # Ensure valid indices and store pixel intensity
    if pixel_array.shape[0] > 269 and pixel_array.shape[1] > 271:
        y.append(int(pixel_array[269, 271]))
    else:
        y.append(0)  # Default value if out of range

# Add colorbar and show image plot
cbar_ax = fig.add_axes([0.92, 0.15, 0.02, 0.7])
fig.colorbar(im, cax=cbar_ax)
plt.show()

# Define the exponential function to fit
def exponential(x, a, b):
    return a * np.exp(b * x)

# Fit the exponential model
params, params_covariance = curve_fit(exponential, x, y, p0=[1, 1])

# Calculate R-squared
y_pred = exponential(x, *params)
ss_res = np.sum((y - y_pred) ** 2)
ss_tot = np.sum((y - np.mean(y)) ** 2)
r_squared = 1 - (ss_res / ss_tot)

# Plot the data
plt.scatter(x, y, label='Data points', color="red")
plt.title("IR-125: Fluorescence (Pixel Intensity) vs Depth")

# Plot the fitted curve
x_model = np.linspace(min(x), max(x), 100)
y_model = exponential(x_model, *params)
plt.plot(x_model, y_model, label='Fit to data')


# Display the fit equation and R-squared value
plt.text(2.0, 1500, f'Equation: $y = {params[0]:.2f} e^{{{params[1]:.2f} x}}$\n$R^2 = {r_squared:.2f}$', fontsize=10, bbox=dict(facecolor='white', alpha=0.5))


plt.xlabel("Depth (mm)")
plt.ylabel("Pixel count")
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

background = 236
file_name = ["1s.txt", "2s.txt", "3s.txt", "4s.txt", "5s.txt", "6s.txt", "7s.txt", "8s.txt", "9s.txt"]

# x values - time in minutes
x = np.array([0, 20, 49, 74, 91, 110, 123, 156, 192])

fig = plt.figure(figsize=(12, 10))
global_min = float('inf')
global_max = float('-inf')

# Load data and find global min/max
for fname in file_name:
    pixel_data = np.loadtxt(fname)
    pixel_array = pixel_data - background
    global_min = min(global_min, pixel_array.min())
    global_max = max(global_max, pixel_array.max())

y = []
y2 = []

# Load data, adjust and display images
for i in range(9):
    pixel_data = np.loadtxt(file_name[i])
    pixel_array = pixel_data - background
    ax = fig.add_subplot(3, 3, i+1)
    im = ax.imshow(pixel_array, cmap='inferno', aspect='equal', vmin=global_min, vmax=global_max)
    ax.set_title(f"{x[i]} mins {int(pixel_array[250, 180])} : {int(pixel_array[250, 330])}")

    # Ensure valid indices and store pixel intensity for both pixel locations
    if pixel_array.shape[0] > 250 and pixel_array.shape[1] > 330:
        y.append(int(pixel_array[250, 180]))
        y2.append(int(pixel_array[250, 330]))
    else:
        y.append(0)  # Default value if out of range
        y2.append(0)  # Default value if out of range

# Convert lists to numpy arrays for element-wise operations
y = np.array(y)
y2 = np.array(y2)

# Normalize y and y2 to their respective first values
if y[0] != 0:  # Prevent division by zero
    y = y / y[0]
if y2[0] != 0:  # Prevent division by zero
    y2 = y2 / y2[0]

# Add colorbar and show image plot
cbar_ax = fig.add_axes([0.92, 0.15, 0.02, 0.7])
fig.colorbar(im, cax=cbar_ax)
plt.show()

# Define the exponential function
def exponential(x, a, b):
    return a * np.exp(b * x)

# Fit the exponential model to normalized y
params, params_covariance = curve_fit(exponential, x, y, p0=[1, 0.01])
y_model = exponential(np.linspace(min(x), max(x), 100), *params)

# Fit the exponential model to normalized y2
params2, params_covariance2 = curve_fit(exponential, x, y2, p0=[1, 0.01])
y2_model = exponential(np.linspace(min(x), max(x), 100), *params2)

# Plot for normalized y
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='red')
plt.plot(np.linspace(min(x), max(x), 100), y_model, label=f'$y = {params[0]:.4f} e^{{ {params[1]:.4f} x}}$')
plt.title("IR-125 Stability")
plt.xlabel("Time (minutes)")
plt.ylabel("Pixel Intensity")
plt.legend(fontsize = 15)
plt.grid(True)
plt.show()

# Plot for normalized y2
plt.figure(figsize=(10, 6))
plt.scatter(x, y2, color='blue')
plt.plot(np.linspace(min(x), max(x), 100), y2_model, label=f'$y = {params2[0]:.4f} e^{{ {params2[1]:.4f} x}}$')
plt.title("IR-765 Stability")
plt.xlabel("Time (minutes)")
plt.ylabel("Pixel Intensity")
plt.legend(fontsize = 15)
plt.grid(True)
plt.show()
```
